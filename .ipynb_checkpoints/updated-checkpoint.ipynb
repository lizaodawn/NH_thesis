```python
#Constructing of the Dataset and assigning UUID

import os

def toposparse(html):
    if not os.path.isfile(html):
        raise FileNotFoundError("File not found: " + html)
    
    data = []
    with open(html, encoding="latin-1") as file:
        soup = BeautifulSoup(file, features="lxml")
        links = soup.find_all("a", {"class": "place"})
        
        for link in links:
            Place_Name = link.contents[0]  # Place name
            ToposText_ID = link.get('about')  # ToposText ID
            Lat = link.get('lat')
            Long = link.get('long')
            Parent = link.find_parent("p")
            Text = Parent.text  # Extract related text
            Reference = Parent.get("id")  # Indicate book, chapter, paragraph
            UUID4 = uuid.uuid4()  # Create a unique ID
            
            data.append({
                'UUID4': UUID4,
                'ToposText_ID': ToposText_ID,
                'Place_Name': Place_Name,
                'Reference': Reference,
                'Lat': Lat,
                'Long': Long,
                'Text': Text
            })
    
    df = pd.DataFrame(data)
    return df
```

## B-2 Text preprocessing for topic modelling

```python
## select all distinct text contents in paragraphs

totaltext = df['Plain_text'].drop_duplicates().str.replace('â\x80\x94', '-', regex=False) ## correct one detected encoding error
totaltext
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

## function for generating tokenlist in paragraphs

def tokenize_paragraphs(texts):
    ## apply word tokenization function of NLTK module to each paragraph
    words = [word_tokenize(text, language='English') for text in texts]
    
    ## Get all punctuation characters
    all_punctuation = string.punctuation

    ## Remove all characters that are not hyphens ("-")
    hyphen_only_punctuation = all_punctuation.replace("-", "")
    
    ## exclude the punctuations
    words = [[word.translate(str.maketrans('', '', hyphen_only_punctuation)) for word in sentence] for sentence in words]
    
    ## set list of stop words to be excluded from the tokenlist with stopwords function in NLTK
    stop_words = set(stopwords.words('English'))
    exclude = ['also', 'like', 'called', 'kind', 'even', 'name', 'made', 'used', 'one', 'day', 'time']
    
    lemmatizer = WordNetLemmatizer()
    
    ## append words of text into a tokenlist
    tokenlistpara = []
    for paragraph in words:
        paragraph_tokens = []
        for token in paragraph:
            if token.lower() not in stop_words and token.lower() not in exclude and len(token.lower()) > 2:
                ## lemmatize the tokens
                lemmatized_token = lemmatizer.lemmatize(token.lower())
                paragraph_tokens.append(lemmatized_token)
        tokenlistpara.append(paragraph_tokens)
    return tokenlistpara
```

```python
## check the bag of words in paragraphs
para = tokenize_paragraphs(totaltext)
para[:2]
```

```python
## function for generating tokenlist in sentences

def tokenize_sentences(texts):
    ## split the text into sentences using sent_tokenize function of NLTK module
    sentences = sent_tokenize(texts, language='English')
    
    ## apply word tokenization function of NLTK module to each sentence
    words = [word_tokenize(sentence, language='English') for sentence in sentences]
    
    ## Get all punctuation characters
    all_punctuation = string.punctuation

    ## Remove all characters that are not hyphens ("-")
    hyphen_only_punctuation = all_punctuation.replace("-", "")
    
    ## exclude the punctuations
    words = [[word.translate(str.maketrans('', '', hyphen_only_punctuation)) for word in sentence] for sentence in words]
    
    ## set list of stop words to be excluded from the tokenlist with stopwords function in NLTK
    stop_words = set(stopwords.words('English'))
    exclude = ['also', 'like', 'called', 'kind', 'even', 'name', 'made', 'used', 'one', 'day', 'time']
    
    lemmatizer = WordNetLemmatizer()
    
    ## append words of text into a tokenlist
    tokenlistsen = []
    for sentence in words:
        sentence_tokens = []
        for token in sentence:
            if token.lower() not in stop_words and token.lower() not in exclude and len(token.lower()) > 2:
                ## lemmatize the tokens
                lemmatized_token = lemmatizer.lemmatize(token.lower())
                sentence_tokens.append(lemmatized_token)
        tokenlistsen.append(sentence_tokens)
    return tokenlistsen
```

```python
## check the bag of words in sentences
totaltextstr = ' '.join(totaltext)
sen = tokenize_sentences(totaltextstr)
sen[:2]
```

```python
## import the necessary modules

from gensim import corpora, models
import gensim

# turn our tokenized documents into a id <-> term dictionary
paradictionary = corpora.Dictionary(para)

# convert tokenized documents into a document-term matrix
paracorpus = [paradictionary.doc2bow(text) for text in para]
```

```python
## do the same process for tokenlist in sentences
sendictionary = corpora.Dictionary(sen)
sencorpus = [sendictionary.doc2bow(text) for text in sen]
```

```python
# generate LDA model with tokenlist in paragraphs
paraldamodel = gensim.models.ldamodel.LdaModel(paracorpus, num_topics=5, id2word = paradictionary, passes=20)
```

```python
## assign 5 topics with 10 for tokenlist in sentences
senldamodel.print_topics(num_topics=5, num_words=10)
```

## D. Initial collocation analysis

```python
## merge the tokens in the bag of sentence as a token list for the whole text 
totaltokenlist = [word for sentence in sen for word in sentence]
totaltokenlist
```

```python
## define a function for generating frequency list of chosen tokenlist

def freqlist (tokenlist):
    freqlist = {}
    for token in tokenlist:
        if token not in freqlist:
            freqlist[token] = 0
        freqlist[token] += 1
    return freqlist
```

### D-1 Bigram of whole text

```python
## generate a bigram of total words in the corpus to find possible patterns

import nltk
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures

# generate a collocation bigram with functions in NLTK
finder = BigramCollocationFinder.from_words(totaltokenlist)
bigram_measures = nltk.collocations.BigramAssocMeasures()

# find the top 20 collocated words in whole text
finder.nbest(bigram_measures.likelihood_ratio, 20)
```

### D-2 Initial collocation analysis on text related to India

As most of the places mentioned were around Mediterranean Sea, India as a rather distanced location yet ranked the fourth among the mentioning counts, I tried extract the texts related to India and did some exploratary analysis.

```python
## select the texts about India
india = df[df['Place_Name'] == 'India']
len(india)
```

```python
## keep only the distinct texts
indiatext = india['Plain_text'].drop_duplicates().str.replace('â\x80\x94', '-', regex=False) ## correct one detected encoding error
len(indiatext)
```

```python
## function for tokenizing given text

def tokenize (text):
    ## apply word tokenization function of NLTK module to the selected text 
    text.apply(word_tokenize, language='English')
    
    ## Get all punctuation characters
    all_punctuation = string.punctuation

    ## Remove all characters that are not hyphens ("-")
    hyphen_only_punctuation = all_punctuation.replace("-", "")
    
    ## exclude the punctuations except for "-"
    word = [line.translate(str.maketrans('', '', hyphen_only_punctuation)).split() for line in text]
    
    ## set list of stop words to be excluded from the tokenlist with stopwords function in NLTK
    stop_words = set(stopwords.words('English'))
    exclude = ['also', 'like', 'called', 'kind', 'even', 'name', 'made', 'used', 'one', 'day', 'time']
    
    lemmatizer = WordNetLemmatizer()
    
    ## append words of text into a tokenlist
    tokenlist = []
    for group in word:
        for token in group:
            if token.lower() not in stop_words and token.lower() not in exclude and len(token.lower()) > 2:
                ## lemmatize the tokens
                lemmatized_token = lemmatizer.lemmatize(token.lower())
                tokenlist.append(lemmatized_token)
    return tokenlist
```

```python
indiatokenlist = tokenize(indiatext)
indiatokenlist
```

```python
## exclude word "india" in the token list

indiatokenlist = [token for token in indiatokenlist if token != 'india']
```

```python
# generate a collocation bigram with functions in NLTK
finder = BigramCollocationFinder.from_words(indiatokenlist)
bigram_measures = nltk.collocations.BigramAssocMeasures()

# find the top 20 collocated words in whole text
finder.nbest(bigram_measures.likelihood_ratio, 20)
```

```python
keyword = "India"
finder = BigramCollocationFinder.from_documents(sen)
finder.apply_word_filter(lambda w: w == keyword.lower())
finder.apply_freq_filter(10)
measures = BigramAssocMeasures()
collocations = finder.score_ngrams(measures.pmi)
collocations[:20]
```

```python
## sort the top 20 words with highest frequency

indiafreqlist = freqlist(indiatokenlist)
sorted(indiafreqlist.items(), key=lambda x:x[1], reverse = True)[:20] 
```

```python
## extract the 5 closest adjectives and verbs about keyword "india" in the whole text
nltk.download('averaged_perceptron_tagger')

## create a ConcordanceIndex object
concordance = nltk.ConcordanceIndex(totaltokenlist)

## define the keyword and context size
keyword = "india"
context_size = 10

## find the offset positions of the keyword in the corpus
offsets = concordance.offsets(keyword)

## extract the adjectives and verbs within the context window
adjectives = set()
verbs = set()
for offset in offsets:
    left_context = totaltokenlist[max(0, offset - context_size):offset]
    right_context = totaltokenlist[offset + 1:min(len(totaltokenlist), offset + context_size + 1)]
    for word in left_context + right_context:
        pos_tag = nltk.pos_tag([word])
        if pos_tag[0][1].startswith("JJ"):
            adjectives.add(word)
        elif pos_tag[0][1].startswith("VB"):
            verbs.add(word)

# Print the results
print("Adjectives:", adjectives)
print("Verbs:", verbs)
```
